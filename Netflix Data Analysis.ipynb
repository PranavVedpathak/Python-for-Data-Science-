```python
import pandas as pd
import matplotlib as plt
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv("Netflix Userbase.csv")
```

```python
df=data.copy()
```

```python
df.head()
```

```python
df.info()
```

```python
# Converting data type of columns in from objext to date time format
df['Join Date']= pd.DatetimeIndex(df['Join Date'])
df['Last Payment Date']= pd.DatetimeIndex(df['Last Payment Date'])
```

```python
df.head()
```

```python
df.info()
```

```python
df['Plan Duration'].value_counts()
```

```python
join_date=pd.to_datetime(df['Join Date'])
last_payment=pd.to_datetime(df['Last Payment Date'])
```

```python
duration=last_payment-join_date
```

```python
df['Duration']=duration
```

```python
df.head()
```

```python
df['Duration'].dt.days//30
```

```python
df['Duration_Month']=df['Duration'].dt.days//30
```

```python
df.head()
```

```python
plt.figure(figsize=(10,4))
plt.title('Months')
sns.lineplot(x=df['User ID'], y=df['Duration_Month'])
plt.ylabel('Revenue In $')
plt.show()
```

```python
# H0- Avg. Customer churn(Time of Hold) duartion in 10 months
# Ha- Avg. Customer churn(Time of Hold) duartion is not 10 months

from scipy.stats import shapiro
stats,p = shapiro(['Duration_Month'])
print('Shapiro Statistical Value=', stats)
print('Shapiro p values=',p)
alpha = 0.05
if p > alpha:
    print('Avg. Customer churn(Time of Hold) duartion in 10 months-H0')
else:
    print('Avg. Customer churn(Time of Hold) duartion is not 10 months-Ha')
```
