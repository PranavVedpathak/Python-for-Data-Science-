```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.datasets import make_regression
```

```python
#Genarate data using make_regression  1000= rows i/p columns=1 noise = data spraeding(Points splitting in graph) somewhere
data= make_regression(n_samples=1000, n_features=1, noise=25)
```

```python
X= data[0]   # X as i/p feature
y= data[1]     # y as target
```

```python
df=pd.DataFrame(X, columns=['Feature1'])
df['Target']=y
```

```python
X_train, X_test, y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=57)
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,y_train)   # Train data of X and y for linear4
```

```python
y_pred=model.predict(X_test)
```

```python
r2_score(y_test, y_pred)     # value> 0.9 overfit model, ideallay approx 0.7-0.8 
```
